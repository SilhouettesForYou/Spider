# 根据指定URL获取当前页面中的视频下载链接

### 定义获取链接的函数

```python
# -*- coding: utf-8 -*-
import re
import base64
import urllib2
import numpy as np
import requests as req
import matplotlib.pyplot as plt
from PIL import Image
from io import BytesIO
from urllib import urlopen  # 用于获取网页
from bs4 import BeautifulSoup  # 用于解析网页


def featch_videos_links(page):
    pattern = re.compile('mp4|mkv|rmvb')
    html = urlopen(page)
    soup = BeautifulSoup(html, 'html.parser', from_encoding="gb18030")
    links = soup.find_all('a')
    for link in links:
        url = link.get('href')
        if pattern.search(url):
            print url
            
# 获取人人影视中的链接，链接格式为ed2k或magnet
# ed2k：用一段字符描述一个KAD网络中的文件
# Magnet：用一段字符描述一个DHT网络中的BT种子文件
def featch_videos_links_zimuzu(page, ed2k=True, magnet=False):
    pattern_ed2k = re.compile('ed2k:.+mp4|mkv|rmvb')
    pattern_magnet = re.compile('magnet:.+')
    if ed2k is True:
        pattern = pattern_ed2k
    if magnet is True:
        pattern = pattern_magnet
    html = urlopen(page)
    soup = BeautifulSoup(html, 'html.parser', from_encoding="gb18030")
    links = soup.find_all('a', href=pattern)
    for link in links:
        url = link.get('href')
        print url

def display_poster(page):
    html = urllib2.urlopen(page)  
    buf = html.read()  
    soup = BeautifulSoup(buf, 'html.parser', from_encoding="gb18030")
    img_urls = soup.find_all('img', src=re.compile('.jpg'))   #正则表达式，匹配图片格式 
    if len(img_urls) == 1:
        fig = plt.figure(figsize=(32, 32))
        ax = fig.add_subplot(111)
        response = req.get(img_urls[0].attrs['src'])
        image = Image.open(BytesIO(response.content))
        ax.imshow(np.array(image))
        ax.set_xticks(())
        ax.set_yticks(())
    else:
        fig, ax = plt.subplots(1, len(img_urls), figsize=(64, 64))
        for i in range(len(img_urls)):
            response = req.get(img_urls[i].attrs['src'])
            image = Image.open(BytesIO(response.content))
            ax[i].imshow(np.array(image))
            ax[i].set_xticks(())
            ax[i].set_yticks(())
    plt.show()
```
