```python
import redis
import sys
import numpy as np


pool = redis.ConnectionPool(host='127.0.0.1', port=6379, db=0)
r = redis.StrictRedis(connection_pool=pool)
```

### Get Keys

```python
[k.decode() for k in r.keys()]
```

### Print Values

```python
import pandas as pd


def load_key(key):
    values = []
    for value in r.smembers(key):
        values.append(value.decode())
    return np.array(values)
    
pd.set_option('max_rows', 10)
```

```python
pd.DataFrame(load_key('NEW_TIMES_VISIT_SET'), columns = ['NEW_TIMES_VISIT_SET'])
```

```python
pd.DataFrame(load_key('CONTENT_UNSUCCESS_SET'), columns = ['CONTENT_UNSUCCESS_SET'])
```

```python
pd.DataFrame(load_key('STORE_FAILED_IMAGE_SET'), columns = ['STORE_FAILED_IMAGE_SET'])
```
